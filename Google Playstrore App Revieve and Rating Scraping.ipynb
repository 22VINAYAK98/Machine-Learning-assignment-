### wordcloud

Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. Word clouds are widely used for analyzing data from social network websites.

```python
!pip install wordcloud
```

```python
from google_play_scraper import Sort, reviews_all
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
comment_words = ' '
stopwords = set(STOPWORDS) # this is used to avoid most commom words linke, a, the , an... etc

import pandas as pd


dict = { 'userName' : [], 
        'Rating' :[] , 
        'Review'  : []}

df = pd.DataFrame.from_dict(dict) # creating data frame 


result = reviews_all(
    'com.king.candycrushsaga',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=1 # defaults to None(means all score)
)

for i in range(len(result)):
    df2 = { 'userName' : result[i]['userName'],
           'Rating' : result[i]['score'],
          'Review' : result[i]['content']}
    df = df.append(df2,ignore_index = True)

for val in df.Review: 

   # making each val to string 
   val = str(val) 

   # split the value 
   tokens = val.split() 

# Converting each token into lowercase so that uppecas and lowercase words not treated as different words 
for i in range(len(tokens)): 
    tokens[i] = tokens[i].lower() 

for words in tokens: 
    comment_words = comment_words + words + ' '


wordcloud = WordCloud(width = 800, height = 800, 
            background_color ='white', 
            stopwords = stopwords, 
            min_font_size = 10).generate(comment_words) 

# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
df.to_csv("rating_1.csv")# saving as csv file for further reference 
```
